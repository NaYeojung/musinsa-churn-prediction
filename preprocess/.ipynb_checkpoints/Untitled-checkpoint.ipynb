```python
import pandas as pd
from scipy import stats
```

```python
df=pd.read_csv('top10000_review_preprocess.csv',encoding='utf-8')
df=df.groupby('hash_id').agg(lambda x: list(x)).reset_index()
```

```python
df['user_id'] = df['user_id'].apply(lambda x: list(set(x)))
df['level'] = df['level'].apply(lambda x: list(set(x)))
df['count'] = df['count'].apply(lambda x: list(set(x)))
df
```

```python
df['count'] = df['count'].apply(lambda x: x[0])
```

```python
df['z_score'] = stats.zscore(df['count'])
threshold = 1
df['is_outlier'] = (df['z_score'].abs() > threshold)
df
```

```python
df=df[df['is_outlier']==False]
```

```python
max(df['count'])
```

```python
min(df['count'])
```

```python
df['latest_review_date'] = df['review_date'].apply(lambda x: max(x))
```

```python
res=[]
for i in df['latest_review_date']:
    res.append(i)
res = pd.to_datetime(res)
```

```python
import datetime
```

```python
month_ = [x.month for x in res]
day_ = [x.day for x in res]
year_ = [x.year for x in res]
df_ddd = pd.DataFrame({'year' : year_, 'month' : month_, 'day' : day_})
df_ddd.groupby(['year', 'month']).count()
```
