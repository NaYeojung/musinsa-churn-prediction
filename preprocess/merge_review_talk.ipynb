```python
import pandas as pd
```

```python
user_df=pd.read_csv('final_user_reviews(551).csv',encoding='utf-8',index_col='Unnamed: 0')
talk_df=pd.read_csv('../crawl/talks/hashid_talks.csv', encoding='utf-8',index_col='Unnamed: 0')
```

```python
grouped_df=user_df.groupby('hash_id').agg(lambda x: list(x)).reset_index()
grouped_df['user_id'] = grouped_df['user_id'].apply(lambda x: list(set(x)))
grouped_df['level'] = grouped_df['level'].apply(lambda x: list(set(x)))
grouped_df['count'] = grouped_df['count'].apply(lambda x: list(set(x)))
```

```python
talk_df.columns=['hash_id','talk_type','talk_id','talk_title','talk_content','talk_date','talk_category','talk_aggregations','talk_thumnail','talk_comment']
```

```python
talk_df=talk_df.groupby('hash_id').agg(lambda x: list(x)).reset_index()
```

```python
merge_df=pd.merge(grouped_df, talk_df, on='hash_id', how='outer')
merge_df
```

```python
merge_df.to_csv('user_review_talk.csv',encoding='utf-8')
```
