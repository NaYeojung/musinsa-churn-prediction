```python
import pandas as pd
import requests as rq
from tqdm import tqdm
```

```python
product_df=pd.read_csv('productCode.csv', encoding='utf-8')
product_df
# category : 상의 1, 아우터 2, 바지 3
```

```python
def fetch_data(code, cate=1):
    url = f"https://goods-detail.musinsa.com/goods/{code}"
    response = rq.get(url)
    product= response.json()
    
    product_data = {}

    data=product.get('data')

    for key, value in data.items():
        product_data[key] = value
    
    return product_data
```

```python
all_fetch_data=[]
for i in tqdm(product_df[product_df['Category']==1]['Goods No']):
    fetch = fetch_data(i)
    all_fetch_data.append(fetch)
    
df = pd.DataFrame(all_fetch_data)
df=df.loc[:,["goodsNo","goodsNm","sex","seasonYear","specialtyCodes","isLimitedQuantity","isBuyForMember","isAppGoods","isOfflineGoods","isFreeReturn","isSale","saleStartDate","saleEndDate","maxUsePointRate","goodsTags","isMusinsaDirectDelivery","isExclusiveMusinsaPay","isExclusiveMusinsaHyundaiCard","isGiftCard","isLimitedDc","isLimitedCoupon","category","goodsPrice","musinsaHyundaiCardEvent","goodsReview","goodsMaterial"]]
df
```

```python
info_cate_1=pd.read_csv('info_cate_1.csv', encoding='utf-8', index_col='Unnamed: 0')
info_cate_1
```

```python
def size_data(code, cate=1):
    url = f"https://goods-detail.musinsa.com/goods/{code}/actual-size"
    response = rq.get(url)
    product= response.json()
    
    product_data = {}

    data=product.get('data')
    size=data.get('sizes')
    for i in range(len(size)):
        size_list=[]
        for j in range(len(size[i]['items'])):
            name=size[i]['items'][j]['name']
            value=str(size[i]['items'][j]['value'])
            size_list.append(name+' '+value)
            
        product_data[size[i]['name']]=size_list

    # for key, value in data.items():
    #     product_data[key] = value
    
    return product_data
```

```python
all_size_list=[]
for i in tqdm(info_cate_1['goodsNo']):
    size=size_data(i)
    all_size_list.append(size)

all_size_list
```

```python
info_cate_1['size']=all_size_list
info_cate_1
```

```python
size_cate_1=info_cate_1.loc[:,['goodsNo','goodsNm','sex','goodsMaterial','size']]
size_cate_1
```
