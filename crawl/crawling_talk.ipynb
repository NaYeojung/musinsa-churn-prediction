```python
import asyncio
import requests as rq
import pandas as pd
from tqdm import tqdm
import nest_asyncio

nest_asyncio.apply()

async def board_data(memberHashId):
    try:
        url = f"https://community.musinsa.com/community/v1/members/{memberHashId}/posts?type=POST&size=30"
        response = rq.get(url)
        board = response.json()
        
        id_li, title_li, content_li, data_li, thumbnail_li = [], [], [], [], []
        
        data = board.get('data')
        for i in data:
            id_li.append(i['id'])
            title_li.append(i['title'])
            content_li.append(i['content'])
            data_li.append(i['createdAt'])
            thumbnail_li.append(i['thumbnail'])
        
        return {"hashid": memberHashId, "type": "post", "postid": id_li, "title": title_li, "content": content_li, "data": data_li, "thumnail": thumbnail_li}
    except Exception as e:
        print(f"Error in board_data: {e}")
        return {}

async def fetch_data(id):
    try:
        url = f"https://community.musinsa.com/community/v1/posts/{id}/comments?size=30&replySize=3&orderBy=OLDEST"
        response = rq.get(url)
        post = response.json()
        data = post.get('data')
        return data
    except Exception as e:
        print(f"Error in fetch_data: {e}")
        return []

async def board_data_com(memberHashId):
    try:
        url = f"https://community.musinsa.com/community/v1/members/{memberHashId}/posts?type=COMMENT&size=30"
        response = rq.get(url)
        board = response.json()
        
        id_li, title_li, content_li, data_li, thumbnail_li = [], [], [], [], []
        
        data = board.get('data')
        for i in data:
            id_li.append(i['id'])
            title_li.append(i['title'])
            content_li.append(i['content'])
            data_li.append(i['createdAt'])
            thumbnail_li.append(i['thumbnail'])
        
        return {"hashid": memberHashId, "type": "comment", "postid": id_li, "title": title_li, "content": content_li, "data": data_li, "thumnail": thumbnail_li}
    except Exception as e:
        print(f"Error in board_data_com: {e}")
        return {}

async def fetch_data_com(id):
    try:
        url = f"https://community.musinsa.com/community/v1/posts/{id}/comments?size=30&replySize=3&orderBy=OLDEST"
        response = rq.get(url)
        post = response.json()
        data = post.get('data')
        return data
    except Exception as e:
        print(f"Error in fetch_data_com: {e}")
        return []

async def final(hashId):
    try:
        bo = await board_data(hashId)
        bo_com = await board_data_com(hashId)

        board_df = pd.DataFrame(bo)
        board_df_com = pd.DataFrame(bo_com)
        
        all_li = []
        for i in range(len(board_df['postid'])):
            fet = await fetch_data(board_df.iloc[i, 2])
            all_li.append(fet)
        board_df['comment'] = all_li
        
        all_li_com = []
        for i in range(len(board_df_com['postid'])):
            fet_com = await fetch_data_com(board_df_com.iloc[i, 2])
            all_li_com.append(fet_com)
        board_df_com['comment'] = all_li_com
        
        return pd.concat([board_df, board_df_com])
    except Exception as e:
        print(f"Error in final: {e}")
        return pd.DataFrame()

async def main():
    df = pd.read_csv('reviews/reviews_2024_01_goods.csv', encoding='utf-8')
    hash_id_list = df['hash_id']

    review_talk = pd.DataFrame(columns=['hashid', 'type', 'postid', 'title', 'content', 'data', 'thumnail', 'comment'])

    for i in tqdm(hash_id_list):
        final_df = await final(i)
        if not final_df.empty:
            review_talk = pd.concat([review_talk, final_df], ignore_index=True)

    review_talk.to_csv('talks/hashid_talks.csv',encoding='uft-8')
    print(review_talk)

asyncio.run(main())
```

```python
# 유저가 댓글 단 글 정보

def board_data_com(memberHashId):
    try:
        url = f"https://community.musinsa.com/community/v1/members/{memberHashId}/posts?type=COMMENT&size=30"
        response = rq.get(url)
        board= response.json()
        
        id_li=[]
        title_li=[]
        content_li=[]
        data_li=[]
        thumbnail_li=[]
        
    
        data=board.get('data')
        for i in data:
            id_li.append(i['id'])
            title_li.append(i['title'])
            content_li.append(i['content'])
            data_li.append(i['createdAt'])
            thumbnail_li.append(i['thumbnail'])
        
    
        return {"hashid":memberHashId,"type":"comment","postid":id_li, "title":title_li, "content":content_li, "data":data_li, "thumnail":thumbnail_li}
    except:
        return

def fetch_data_com(id):
    try:
        url = f"https://community.musinsa.com/community/v1/posts/{id}/comments?size=30&replySize=3&orderBy=OLDEST"
        response = rq.get(url)
        post= response.json()
        
        comment_li=[]
    
        data=post.get('data')
        # for i in data:
        #     comment_li.append([i['content'],i['replies']])
        
        return data
    except:
        return
```

```python
def final(hashId):
    # 게시글 제목 & content
    bo=board_data(hashId)
    bo_com=board_data_com(hashId)
    board_df=pd.DataFrame(bo)
    board_df_com=pd.DataFrame(bo_com)
    
    # 게시글 하나씩 들어가서 댓글
    all_li=[]
    try:
        for i in tqdm(range(len(board_df['postid']))):
            fet=fetch_data(board_df.iloc[i,2])
            all_li.append(fet)
        board_df['comment']=all_li
        
        all_li_com=[]
        for i in tqdm(range(len(board_df_com['postid']))):
            fet_com=fetch_data_com(board_df_com.iloc[i,2])
            all_li_com.append(fet_com)
        board_df_com['comment']=all_li_com
        
        # post랑 comment 합침
        return pd.concat([board_df,board_df_com])
    except:
        return pd.DataFrame()
```
