```python
import pandas as pd
```

```python
df=pd.read_csv('hashid_talks.csv',encoding='utf-8',index_col='Unnamed: 0')
df
```

```python
count_df = df.groupby(['hashid', 'type']).size().unstack(fill_value=0)

# 새로운 열 'post_count'와 'commend_count' 생성
count_df['post_count'] = count_df.get('post', 0)
count_df['comment_count'] = count_df.get('comment', 0)

# 필요한 열만 선택
result_df = count_df[['post_count', 'comment_count']].reset_index()
result_df
```

```python
result_df.to_csv('user_post_info.csv',encoding='utf-8')
```
